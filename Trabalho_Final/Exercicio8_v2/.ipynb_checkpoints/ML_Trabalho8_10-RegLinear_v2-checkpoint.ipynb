## 9) Ajuste de uma Regressão Linear aos dados para predizer V2 usando cada uma das features do dataset:

### a) Funcao para calcular o Erro Medio Quadratico de cada variável em relação a V2 e imprimir cada MSE, além de indicar qual variável possui o menor MSE 

```python
def mse_features(): 
    
    features = ['V1','V3','V4','V5','V6','V7','V8','V9','V10']
    meta_col = ['V2']
    
    Y_predtr = []
    Y_predte = []
    mse_h_tr = []
    mse_h_te = []

    for i in range(len(features)):
        
        X = data[features[i]].values.reshape(len(data),1)
        y = data[meta_col].values
        split_test_size = 0.30 # 30% da amostra é para treino
        
        XNew = StandardScaler().fit_transform(X)
        Xold = X
        X = XNew

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split_test_size, random_state=25)

        # Training
        
        print('\n','\033[1;92m TRAINING for {} index={}: \033[0;92m'.format(features[i], i))
        lr_model = linear_model.LinearRegression()
        lr_model.fit(X_train, y_train.ravel()) #training the algorithm
        y_predtr = lr_model.predict(X_train)
        print("\033[1;31m Mean squared error from Training Step: \033[0;0m %.2f " % (mean_squared_error(y_train, y_predtr)))
        Y_predtr.append(y_predtr)
        mse_h_tr.append(mean_squared_error(y_train, y_predtr))

        # Test
        
        print('\033[1;92m TEST for {} index={}: \033[0;92m'.format(features[i], i))
        y_predte = lr_model.predict(X_test)
        print("\033[1;31m Mean squared error from Test Step: \033[0;0m %.2f " % (mean_squared_error(y_test, y_predte)),'\n')
        Y_predte.append(y_predte)
        mse_h_te.append(mean_squared_error(y_test, y_predte))

        
    return Y_predtr, Y_predte, mse_h_tr, mse_h_te

y_predtr, y_predte, mse_h_tr, mse_h_te = mse_features()
print('\n', "\033[1;96m Menor MSE = %.1f. \033[0;96m" % np.min(mse_h_te))
print('\n', "\033[1;96m O menor MSE encontrado é da variável {} \033[0;96m".format(features[int(np.argmin(mse_h_te))]))
```

### b) Retorne um gráfico com a reta ajustada aos dados para cada feature e diga se ela é uma variável boa ou não para predizer V2.

```python
import random

features = ['V1','V3','V4','V5','V6','V7','V8','V9','V10']

# Regression plot using seaborn.
fig = plt.figure(figsize=(10,7))

for i in range(len(features)):
    r = lambda: random.randint(0,255)
    c = '#%02X%02X%02X' % (r(),r(),r())
    sb.regplot(x=y_test,y=y_predte[i],color=c)

# Legend, title and labels.
plt.legend(labels=['V1','V3','V4','V5','V6','V7','V8','V9','V10'], bbox_to_anchor=(0.25, 0.95))
plt.title('Relação entre as demais Variáveis e V2', size=16)
plt.xlabel('Features', size=14)
plt.ylabel('V2', size=14);
plt.show()
```
